### 문자열 인덱싱(가리키기)과 슬라이싱(잘라내기)

*   대괄호[ ]를 사용해 객체의 특정 요소(들)을 가리키고 잘라내기

    **"파이썬은 요소를 가리킬때 0부터 숫자를 센다"**

*   인덱싱: 문자열을 나열했을 때 각 문자의 위치에 번호를 매겨 특정한 값을 뽑아낼때 사용(예, text[3])

*   슬라이싱: 문자열의 문자들의 범위를 추출할 때 사용(예, [n1:n2] = n1이상 n2미만)


```python
# 1. 문자열 인덱싱: 문자열 요소 가리키기
b = 'KHU'
print(b[-1])

# docstring 문자열 예제
s_2 = '''I don't know what we've done
되돌아가긴 싫어, もう知っている
Don't know what we've been sold
見つけられるよ, so it's sure (come on)
'''
# s_2 문자열의 첫 번째 문자를 가리키기
s_2[0]

# s_2 문자열의 두 번째 문자  (공백 예상)
s_2[1]

# s_2 문자열의 마지막 문자를 가리키기(줄바꿈 표시(\n) 예상)
s_2[-1]

# s_2 문자열의 단어 길이 구하기: len()함수 이용하기
len(s_2) #111 단어들(공백포함) 요소
```

```python
# 2. 문자열 슬라이싱: 문자열 요소 잘라내기

s_2 = '''I don't know what we've done
되돌아가긴 싫어, もう知っている
Don't know what we've been sold
見つけられるよ, so it's sure (come on)
'''


# 처음부터 4번째까지 잘라내기
s_2[0]+s_2[1]+s_2[2]+s_2[3]

s_2[0:4]  # 0 이상부터 4 미만까지 잘라내기

s_2[:10]  # 0 이상부터 10 미만까지 잘라내기
```

### 문자열 관련 주요 내장함수

### 기본 if - else 문

```python
# if + 조건:
#  members 리스트에 "하니"라는 이름이 포함되어 있는지를 확인한 후, 그 결과에 따라 다른 메시지를 출력하는 코드

members = ["하니", "해린", "민지", "혜인", "다니엘"]

if "하니" in members: #in 연산자는 리스트, 문자열 등에서 특정 요소의 존재 여부를 확인하는 데 사용
    print("하니는 뉴진스 멤버입니다")
else:
    print("하니는 뉴진스 멤버가 아닙니다")
```

### if - elif - else 문

```python
# if + 조건1:
# elif + 조건2:

# 변수 p1과 p2에 각각 "하니"와 "혜인"이라는 이름을 할당하고,
# 하니_생년과 혜인_생년에 각각 그들의 출생 연도를 할당한 후,
# 비교연산자(<, >, >=, <=, ==, !=)를 이용해 이 두 사람의 나이를 비교하여 서로의 관계를 출력하는 조건문을 작성


# 변수 선언
p1, p2 = "하니", "혜인"

# 나이 정보 선언
하니_출생연도, 혜인_출생연도 = 2004, 2008

# 조건문
if 하니_출생연도 == 혜인_출생연도:
    print(f"{p1}와 {p2}는 동갑입니다")
elif 하니_출생연도 < 혜인_출생연도:
    print(f"{p1}가 {p2}의 언니입니다")
else:
    print(f"{p1}가 {p2}의 동생입니다")
```

## 제어문(2) - while문(반복문)


*   while문: 특정 조건이 참(True)인 동안, 지정된 블록의 문장들을 반복해서 실행

*   조건이 거짓이 될 때까지 계속 반복 (예, 나무를 10번찍을 때까지 반복) - 반복이 언제 끝나는지를 제어

*   만약 조건이 항상 참인 경우 무한히 실행될 수 있음(무한 루프) - 조건을 변경해서 빠져 나오려면 break를 코드내에 삽입


image.png

```python
# 예제 1: 조건이 참인 동안 반복 실행
# while 루프를 사용하여 x가 10보다 작은 동안 1씩 증가
# x가 10이 되면 while 조건이 거짓이 됨. 루프가 종료

x=1
while x < 10:  # 10보다 작은 동안 반복 (9까지)
    x += 1      # x가 10보다 작은 동안에는 계속 1을 더하다가, 10 되자 while 조건에 걸려 실행 중단.
    print(x)    # 2부터 ~10까지 출력
```

```python
# 예제 2: 조건이 참인 동안 반복 실행: 무한루프와 강제종료

x = 1
while True:  # 무한 루프
    print(x)
    x += 1
    break                # break가 없고 멈추려면, 런타임 > 실행중단 혹은 혹은 ctrl + c.

```

## 제어문(3) -for문(반복문)

*   for문: 주어진 요소들 전체에 대해서 [코드 블록]이 반복적으로 실행 (예, **리스트 모든 요소**에 대해 반복문이 동작할 수 있도록 제어)

*   in 뒤 리스트에서 지정된 요소를 순서대로 실행

*   for-in- 방식으로 표기

*   리스트 뿐만 아니라 튜플과 문자열에도 적용됨

image.png

```python
# 예제 1
# members 리스트에 포함된 각 멤버의 이름을 인덱싱으로 하나씩 접근해서 출력하는 예제
members = ["하니", "해린", "민지", "혜인", "다니엘"]

print(members[0])
print(members[1])
print(members[2])
print(members[3])
print(members[4])
```

```python
# 예제 2
# for 루프를 사용하여 members 리스트에 포함된 각 멤버의 이름을 순회하며, 각 멤버 이름을 출력하는 예제

members = ["하니", "해린", "민지", "혜인", "다니엘"]
for i in members: # 여기에서 i는 순회된 멤버들의 이름을 각각 할당시키기위해 코더가 생성한 변수
    print(i)      # i는 문자열: i = "하니","해린","민지","혜인","다니엘"
```

```python
 # 예제 3
 # for 루프를 사용하여 members 리스트에 포함된 각 멤버의 이름을 순회하며, 각 멤버가 뉴진스 멤버라는 문장을 출력하는 예제

members = ["하니", "해린", "민지", "혜인", "다니엘"]
for i in members:  # 여기에서 i는 순회된 멤버들의 이름을 각각 할당시키기위해 코더가 생성한 변수
    print(f"{i}는 뉴진스 멤버입니다")
```

```python
# 예제 4
# for 루프와 if-else 조건문을 사용하여 시험 리스트에 포함된 각 점수를 평가하고,
# 그 점수가 80점 이상이면 "귀가"를, 그렇지 않으면 "재시"를 출력하라

시험점수 = [95, 55, 71, 83]

for 개별점수 in 시험점수: # 여기에서 '개별점수'는 순회된 시험점수들의 각각을 할당시키기위해 코더가 생성한 변수
    if 개별점수 >= 80:
        print("귀가")
    else:
        print("재시")
```

```python
# 예제 5 (중요함: 예제 8과 비교)
# for문을 사용해서 '시험점수' 리스트를 평가하여 "귀가" 또는 "재시"를 새로운 '성적평가' 리스트로 결과를 반환하라
# 힌트: 새로운 '성적평가'리스트를 생성하고, 각각 시험점수에 매칭되는 "귀가" 또는 "재시"를 그 리스트에 추가할 append() 함수를 사용하라


시험점수 = [95, 55, 71, 83]
성적평가 = []

for 개별점수 in 시험점수:# 여기에서 '개별점수'는 순회된 시험점수들의 각각을 할당시키기위해 코더가 생성한 변수
    if 개별점수 >= 80:
        성적평가.append("귀가")
    else:
        성적평가.append("재시")

# 평가 리스트에 결과가 담깁니다. 예상 정적평가 리스트는 ['귀가', '재시', '재시', '귀가']
print(성적평가)

```

### 예외 처리(try/except)
- **예외(Exception)**: 코드 실행 중에 발생하는 오류
- **프로그램이 중간에 멈추지 않도록, 에러를 처리하는 방법이 필요**
- `try` 블록에 에러가 발생할 수 있는 코드를 작성
- `except` 블록에 에러가 발생했을 때 실행할 코드를 작성

```python
#"뉴진스 멤버 확인 프로그램 (예외 처리 포함)"
# 리스트에서 특정 이름이 있는지 확인하면서,예외가 발생해도 프로그램이 멈추지 않고 계속 실행되도록 만든 예제

members = ["하니", "해린", "민지", "혜인", "다니엘"]

try: # 블록 안에서 코드를 실행 - 실행중 오류가 발생하면 멈추지 않고 excpet 블로으로 이동
    name = input("이름을 입력하세요: ")
    if name in members:
        print(f"{name}는 뉴진스 멤버입니다.")
    else:
        print(f"{name}는 뉴진스 멤버가 아닙니다.")
except Exception as e: # 모든 에러를 잡아 변수 e에 담는다는 뜻
    print("에러가 발생했습니다:", e)

```

# 함수(function)

*   입력값을 받아 어떤 일을 실행한 후 그 결과물을 내어놓는 도구

*   예) y = 2x + 1 함수는 x라는 input, y라는 output을 내놓음

*   예) 과일(입력값) > 믹서기(함수) > 주스(결과값)

*   함수는 입력 --> 작업 -> 출력.

*   반복해서 코드를 재사용.

*   사용형식에서 콜론과 들여쓰기 중요함

*   아래 함수의 구조 참조

image.png

*   def: 함수 정의할때 사용하는 키워드

*   매개변수(parameter): 함수에 입력으로 전달될 값을 받는 변수

*   인자(arguments): 함수를 호출할 때 전달하는 입력값

```python
# 예제1: 함수의 기본 구조 1

def add(x, y): # def는 함수 정의할때 사용하는 키워드, add는 함수명,  x, y는 매개변수(입력변수), 초기값 설정가능
    result = x + y # 실행문: result는 결과값
    return result  # 결과값을 반환하라

```

# 클래스(class)

*   클래스는 객체지향 프로그래밍에서 필수적인 개념

*   클래스(Class): 똑같은 것을 계속 만들어 낼 수 있는 설계 도면. 같은 구조와 기능을 가진 객체를 반복해서 만들수 있는 틀(붕어빵 틀에 비유)

*   객체(object) 또는 인스턴스(instance): 클래스라는 틀로 찍어낸 실제 붕어빵. 예를 들어 "붕어빵 틀(클래스)로 "붕어빵(객체)"을 여러 개 만들 수 있음. 다양한 의자들도 인스턴스임.

*   메서드(method): 클래스 안에 구현된 함수. 붕어빵을 만들때 반죽 붓고, 굽고, 꺼내는 과정처럼 메서드는 객체가 수행할 수 있는 동작들을 정의함

*   생성자(constructor): 객체가 생성될 때 자동으로 호출되는 메서드. 이 생성자를 통해 객체에 필요한 초깃값을 설정할 수 있음

*  클래스: (원형) 설계도
*  객체(인스턴스): 설계도로 만든 실제 제품
*  메서드: 객체가 할 수 있는 동작
*  생성자(__init__): 객체 생성시 초기값 설정

```python
# 예제1: 간단한 계산기 클래스를 정의한 예제 > 계산기 설계도
# 이 코드는 "Calculation"이라는 클래스를 정의하고 있음
# 이 클래스는 두 숫자를 입력받아 더하기, 빼기, 곱하기, 나누기 같은 기본적인 계산을 할 수 있는 기능을 가지고 있음

class Calculation:                      # 클래스는 대문자로("C"aculation), 객체는 소문자로
    def __init__(self, n1, n2):         #  __init__(self, n1, n2)는 생성자(constructor) 표시.
        self.n1 = n1                    # 아래에 객체가 생성될 때 이 생성자가 자동으로 호출됨. 객체 속성의 초기화작업 담당
        self.n2 = n2                    # 객체가 생성될 때, 전달된 두 개의 값 n1과 n2를 해당 객체의 속성 self.n1과 self.n2에 저장하여 초기화합니다.
                                        # self는 아래 생성될 객체 자신을 가리킴, 이를 통해 클래스 속성에 접근할 수 있음

    def add(self):                      # 객체가 수행할 동작인 2개의 함수(메서드) 정의: 더하기, 곱하기
        result = self.n1 + self.n2
        return result

    def mul(self):
        result = self.n1 * self.n2
        return result
```

* 객체(인스턴스) 생성하기 = 설계도로 만든 제품 만들기
* 위에서 정의한 클래스를 사용해서 특정한 객체 생성
* 예) 붕어빵 틀에 반죽을 부어 붕어빵 하나를 만드는 것
* 클래스를 이용해 객체(인스턴스)를 만들 때는, 보통 변수에 저장해서 사용

```python
# 예제2: 클래스를 이용해서 특정 속성(8. 5)을 가진 객체(cal1) 생성하기
cal1 = Calculation(8, 5) # 이때 클래스의 생성자인 __init__ 메서드를 자동으로 호출

# # cal1 객체인 속성인 cal2.n1와 cal2.n2는?
cal1.n1 # 객체의 속성 8
cal1.n2 # 객체의 속성 5

# <클래스, 객체, 속성, 메서드에 대한 설명>
# Calculation 클래스라는 설계도를 사용해 'cal1'이라는 계산기 객체를 하나 만듬
# 'cal1' 객체는 8과 5를 가지고 태어나며(=객체의 속성)
# 클래스의 _init__ 메서드를 자동으로 호출하면서 n1과 n2에 각각 8과 5가 전달되며 이 값들은 self.n1과 self.n2에 저장
# 즉, n1과 n2를 받아서, 객체(cal)의 속성인 self.n1과 self.n2를 초기화하는 역할
# 클래스의 메서드인 add()과 mul()같은 동작을 수행할 수 있음
```

```python
# 예제3: 클래스를 이용해서 특정 속성(8. 5)을 가진 객체(cal1) 생성하기
cal1 = Calculation(8, 5)

# cal1 객체가 수행할 메서드 호출
# 호출된 객체(cal1)가 수행할 더하기,곱하기 동작 메서드의 결과는?
cal1.add()         # 13
cal1.mul()         # 40

```

```python
# 예제4(집에서 해볼것)
# 클래스를 이용해서 특정 속성(10, 2)을 가진 객체(cal2) 생성하기
cal2 = Calculation(10, 2)

# cal2 객체인 속성인 cal2.n1와 cal2.n2는?
cal2.n1
cal2.n2

# cal2 객체가 수행할 메서드 호출:
cal2.add()         # ?
cal2.mul()         # ?
```

# 파일 쓰기와 불러오기(txt 파일)

## 1. 파일 다루기 개요

*   텍스트 파일을 쓰고 저장하는 방법과, 저장된 파일을 불러와 읽는 방법

```python
# 파일 불러와서 읽기
with open("/content/drive/MyDrive/mypython/test.txt", "r") as f: # open() 함수로 "test.txt" 파일을 **읽기 모드("r")**로 염.f는 열린 파일을 나타내는 파일 객체
    mytext1 = f.read() #f.read()를 통해 test.text 파일(=f 파일객체)의 모든 내용을 읽어와 mytext1 변수에 저장

print(mytext1) #저장된 텍스트를 출력
```

* 즉, 데이터프레임은 파이썬 자료형과 CSV나 엑셀과 같은 스프레드 시트 파일 사이를 연결해주는 다리이자, 파이썬에서 표 데이터를 다루는 핵심 도구임

```python
# Yes24 베스트셀러 페이지에서 책 제목, 저자, 출간년도, 출판사를 추출해 Pandas 데이터프레임으로 정리하는 코드

import pandas as pd

# 데이터 저장을 위한 리스트 초기화
data = []

# 책 정보 리스트에서 데이터 추출 후 리스트에 저장
for book in books:
    # 서적 이름 추출
    title_tag = book.find('a', class_='gd_name')
    title = title_tag.get_text(strip=True) if title_tag else "N/A"

    # 저자명 추출
    author_tag = book.find('span', class_='authPub info_auth')
    author = author_tag.find('a').get_text(strip=True) if author_tag and author_tag.find('a') else "N/A"

    # 출간년도 추출
    publication_date_tag = book.find('span', class_='authPub info_date')
    publication_date = publication_date_tag.get_text(strip=True) if publication_date_tag else "N/A"

    # 출판사명 추출
    publisher_tag = book.find('span', class_='authPub info_pub')
    publisher = publisher_tag.find('a').get_text(strip=True) if publisher_tag and publisher_tag.find('a') else "N/A"

    # 데이터를 리스트에 추가
    data.append([title, author, publication_date, publisher])

# Pandas 데이터프레임 생성
df = pd.DataFrame(data, columns=['서적 이름', '저자', '출간년도', '출판사'])

# 데이터프레임 앞 부분만 미리 확인가능
df.head()

# 전체데이터프레임 출력
print(df)
```
